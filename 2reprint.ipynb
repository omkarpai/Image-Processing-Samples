### Aim: To read an image in python in various color modes and display its dimensions.

### Part A Color Image

```python
import numpy as np 
import cv2
img = cv2.imread('doomguy256hd.png',1)#reading the image in color
#print(img)#displaying pixel values
print("\n")
print("Number of rows =",img.shape[0])#displaying dimensions
print("Number of colums =",img.shape[1])
print("Dimension of image =",img.shape[0],"x",img.shape[1])
try:
    print("Number of channels =",img.shape[2])
except:
    print("Image is grayscale")
    
dtype = img.dtype
if(dtype =="uint8"):
    sizedisk = ((img.size*8)/(1024*8))#disk size of 8 bit image
    print("Size of image on disk is",sizedisk ,"KB without compression")
```

### Part B Grayscale image

```python
import numpy as np 
import cv2
img = cv2.imread('doomguy256hd.png',0)#reading the image in grayscale
cv2.imwrite("gs_doomguy256.png", img)
print("\n")
print("Number of rows =",img.shape[0])#displaying dimensions
print("Number of colums =",img.shape[1])
print("Dimension of image =",img.shape[0],"x",img.shape[1])
try:
    print("Number of channels =",img.shape[2])
except:
    print("Image is grayscale")
    

dtype = img.dtype
if(dtype =="uint8"):
    sizedisk = ((img.size*8)/(1024*8))
    print("Size of image on disk is",sizedisk ,"KB without compression")
```

### Part C Black & White image

```python
import numpy as np 
import cv2
img = cv2.imread('doomguy256hd.png',0)#reading the image in grayscale
(thresh, imgbw) = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)
cv2.imwrite('bw_doomguy256hd.png', imgbw)
cv2.imshow("bw_doomguy256hd",imgbw)#displaying the image
print("\n")
print("Number of rows =",img.shape[0])#displaying dimensions
print("Number of colums =",img.shape[1])
print("Dimension of image =",img.shape[0],"x",img.shape[1])
try:
    print("Number of channels =",img.shape[2])
except:
    print("Image is grayscale")
    
dtype = img.dtype
if(dtype =="uint8"):
    sizedisk = ((img.size*8)/(1024*8))
    print("Size of image on disk is",sizedisk ,"KB without compression")
```
